```python
import pandas as pd
import os
```

```python
florida_file = os.path.join("Resources", "fl_ev_registrations_public.csv")

florida_ev = pd.read_csv(florida_file)
florida_ev.head()
```

```python
florida_ev['State'] = 'FL'
florida_ev.head()
```

```python
reduced_florida_ev = florida_ev[['State', 'Vehicle Name']]
reduced_florida_ev.head()
```

```python
reduced_florida_ev[['Make', 'Model']] = reduced_florida_ev['Vehicle Name'].str.split(' ', 1, expand=True)
reduced_florida_ev.head()
```

```python
final_florida_ev = reduced_florida_ev[['Model', 'Make', 'State']]
final_florida_ev.head()                                        
```

```python
total_models = final_florida_ev.groupby('Model')['Make'].count()
total_models.head()
```

```python
summary_florida_ev = final_florida_ev.groupby('Model').first()

summary_florida_ev["Total Models"] = total_models
```
