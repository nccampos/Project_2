```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Note: Please download the csv from the source website and place it within the Resources folder.

#load csv file into juypter notebook
ca_file = pd.read_csv("Resources/ca_ev_registrations_public.csv")

#drop duplicate vehicle IDs
ca_file["Vehicle ID"].drop_duplicates()

#remove unneccessary columns
filtered_data = ca_file[["State Abbreviation", "Geography", "Vehicle Name"]]

#spilt vehicle name column into make and model
filtered_data[["Make", "Model"]] = filtered_data["Vehicle Name"].str.split(' ', n=1, expand=True)

#filter to final columns and change state column name
ca_df= filtered_data[["State Abbreviation", "Make", "Model"]]

ca_df = ca_df.rename(columns={"State Abbreviation":"State"})

ca_df.head()
```

```python
#create a df for total models in the states
a=ca_df.groupby('Model')['Model'].count()

#create new df for vehicle count
ca_count_df = pd.DataFrame(data=a)

#rename column as "count", as for the number of those vehicles in the state
ca_count_df = ca_count_df.rename(columns={"Model":"count"})

#add a state column to the df
ca_count_df["State"]= "CA"
```

Database Connection

```python
#First create the db schema on Postgres pg admin 4
#DB Name "ev_count_db"
#Table Names ="CA" , "CO", etc..

connection_string = "postgres:postgres@localhost:5432/ev_count_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Load DataFrames into database
ca_count_df.to_sql(name='ca', con=engine, if_exists='append', index=True)
co_count_df.to_sql(name='co', con=engine, if_exists='append', index=True)
fl_count_df.to_sql(name='fl', con=engine, if_exists='append', index=True)
mn_count_df.to_sql(name='mn', con=engine, if_exists='append', index=True)
nj_count_df.to_sql(name='nj', con=engine, if_exists='append', index=True)
ny_count_df.to_sql(name='ny', con=engine, if_exists='append', index=True)
tn_count_df.to_sql(name='tn', con=engine, if_exists='append', index=True)
tx_count_df.to_sql(name='tx', con=engine, if_exists='append', index=True)
va_count_df.to_sql(name='va', con=engine, if_exists='append', index=True)
wa_count_df.to_sql(name='wa', con=engine, if_exists='append', index=True)
```
