```python
import pandas as pd
from sqlalchemy import create_engine
import os
```

```python
speed_path = os.path.join("Resources","Quickestelectriccars_EVDatabase.csv")
speed_df = pd.read_csv(speed_path)
```

```python
speed_df.head()
```

```python
reduced_speed_df = speed_df.drop(columns = ["Subtitle","Drive","NumberofSeats","PriceinGermany","PriceinUK"])
reduced_speed_df.head()
```

```python
reduced_speed_df[['Name', 'Make']] = reduced_speed_df['Name'].str.split(' ', 1, expand=True)
reduced_speed_df.head()
```

```python
final_speed_df = reduced_speed_df[['Name', 'Make', 'Acceleration','TopSpeed', 'Range','FastChargeSpeed']]
final_speed_df.head()
```

```python
#First create the db schema on Postgres pg admin 4
#DB Name "speed_db"
#Table Names ="EVMODELS" 

connection_string = "postgres:postgres@localhost:5432/speed_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Load DataFrames into database
final_speed_df.to_sql(name='EVMODELS', con=engine, if_exists='append', index=True)
```

```python
#check to confirm data has been loaded properly to table and db
pd.read_sql_query('select * from EVMODELS', con=engine).head()
```
