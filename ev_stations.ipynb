```python
#Dependencies
import pandas as pd

evloc_df = pd.read_csv('Resources/ev_stations_v1.csv')
#evloc_df.columns
#evloc_df.head()
```

```python
# Choosing which columns to keep
evloc_df = evloc_df[['Station Name', 'Street Address', 'City', 'State', 'ZIP', 'Access Code', 'EV Network', 'EV Connector Types']]

#evloc_df['Fuel Type Code'].unique()
#evloc_df['State'].unique()

#output:
#array(['CA', 'VT', 'WA', 'OR', 'IL', 'ID', 'TX', 'TN', 'NY', 'FL', 'WI',
#       'IA', 'MN', 'AR', 'AZ', 'CT', 'MD', 'NJ', 'SC', 'MI', 'OH', 'WV',
#       'MO', 'UT', 'KS', 'MA', 'CO', 'IN', 'LA', 'NC', 'NH', 'VA', 'AL',
#       'GA', 'HI', 'NV', 'DC', 'PA', 'RI', 'OK', 'ME', 'KY', 'NE', 'MS',
#       'SD', 'DE', 'NM', 'MT', 'ND', 'WY', 'AK', 'PR', 'ON'], dtype=object)
```

```python
# Filtering dataframe by specific states
tten = ['WA','CA','NY','CO','TX','FL','NJ','TN','MN','VA']

# Top 10 States with EVs registered, according to Atlus EV Hub and Open Vehicle Registration Initiative:
# Washington, California, New York, Colorado, Texas, Florida, New Jersey, Tennessee, Minnesota, Virginia

tenloc = evloc_df[evloc_df.State.isin(tten)]
#tenloc.head()
```

```python
# Renaming columns
rename = {'Station Name':'station_name', 
            'Street Address':'address', 
            'City':'city', 
            'State':'state', 
            'ZIP':'zip',
            'Access Code':'access', 
            'EV Network':'ev_network', 
            'EV Connector Types':'connector_type'}

tenloc = tenloc.rename(mapper=rename, axis=1)
```

```python
# Resetting index
tenloc = tenloc.reset_index()
```

```python
# Deleting extra index column
del tenloc['index']
```

```python
# Exporting to csv
tenloc.to_csv('ev_stations_statesorted.csv')

# .to_html() if needed
```
