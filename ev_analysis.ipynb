```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as matplot
import seaborn as sns
%matplotlib inline
```

```python
path_normal='/Resources/ElectricCarData_Clean.csv'
```

```python
car_data=pd.read_csv(path_normal)
car_data.head()
```

Finding unique values

```python
print(f'Brand {car_data.Brand.unique()}')
print(f'Rapid Charge {car_data.RapidCharge.unique()}')
print(f'Power Train {car_data.PowerTrain.unique()}')
print(f'Plug Type {car_data.PlugType.unique()}')
print(f'Body Style {car_data.BodyStyle.unique()}')
print(f'Segment {car_data.Segment.unique()}')
```

checking for null data

```python
car_data.isnull().sum()
```

Which car has fastest acceleration 0=100

```python
plt.figure(figsize=(50,20))
sns.barplot(x='Brand',y='AccelSec',data=car_data)
plt.title('Brand Vs AcceleSec ')
```

Fastest car is in the brand name "Lucid"

Which has the highest efficiency

```python
plt.figure(figsize=(50,20))
sns.barplot(x='Brand',y='Efficiency_WhKm',data=car_data)
plt.title('Brand Vs Effienciency ')
```

Byton have the highest efficiency

Does a difference in power train effect the range, top speed, efficiency?

```python
sns.catplot(x='PowerTrain',y='Range_Km',data=car_data)
plt.title('PowerTrain Vs Range')
sns.catplot(x='PowerTrain',y='Efficiency_WhKm',data=car_data)
plt.title('Power Train Vs Efficiency')

sns.catplot(x='PowerTrain',y='TopSpeed_KmH',data=car_data)
plt.title('PowerTrain Vs Top Speed')
```

Which Manufacturer has the most number of vehicles

```python
Most_car=car_data.groupby('Brand').count()
print(Most_car['Model'].sort_values(ascending = False))
```

Tesla produces most number of vehicles

How does price relate to rapid charging

```python
sns.swarmplot(x='PriceEuro',y='RapidCharge',data=car_data)
plt.title('Price Vs Rapid_Charge')
```
